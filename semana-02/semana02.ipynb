### Operação com arrays

```python
# operacções aritmétricas
#som
array1 = np.array([1,2,3,4])
array2 = np.array([4,5,6,7])
array_aux = array1 + array2
print(array_aux)

#subtração
array1 = np.array([1,2,3,4])
array2 = np.array([4,5,6,7])
array_aux = array1 - array2
print(array_aux)

#produto
array1 = np.array([1,2,3,4])
array2 = np.array([4,5,6,7])
array_aux = array1 * array2
print(array_aux)

#divisão
array1 = np.array([1,2,3,4])
array2 = np.array([4,5,6,7])
array_aux = array1 / array2
print(array_aux)
```

```python
# operação com escalar
#soma
array_aux = array2 + 5
print(array_aux)

#subtração
array_aux = array2 - 5
print(array_aux)

#produto
array_aux = array2 * 5
print(array_aux)

#divisão
array_aux = array2 / 5
print(array_aux)
```

```python
# broadcasting
array1 = np.array([[10,11,13],[15,16,17]])
print(array1)
array2 = np.array([10,20,30])
print(array2)

soma = array1 + array2
print(soma)
```

### Funções de agregação

```python
# sum
soma = array1.sum()
print(soma)
```

```python
# mean - media aritmétrica
media = array1.mean()
print(media)
```

```python
# std - calcula desvio padrão dos elementos
dp = array2.std()
print(dp)
```

```python
# min
valor_min = array1.min()
print(valor_min)
```

```python
#max
valor_max = array2.max()
print(valor_max)
```

```python
# por linha e coluna
min_linha = array1.min(axis=1)
max_coluna = array1.max(axis=0)
print(min_linha, max_coluna)
```

```python
# argmax
indice_maior_valor = array1.argmax()
print(indice_maior_valor)
```

### Operações lógicas e comparações

```python
# operadores lógicos
array_bool = np.array([True, False, True])
array_bool2 = np.array([False, False,True])

#and
array_and = array_bool & array_bool2
print(array_and)

#or
array_or = array_bool | array_bool2
print(array_or)
```

```python
# filtragem
dados = np.array([1,2,3,4,5,6,7])

filtro = dados < 5
print(filtro)
```

### Indexação e fatiamento

```python
# fatiamento
# acessando elementos
array = np.arange(10, 60, 10)
print(array)

print(array[0])
print(array[2])

#matriz
matriz = np.array([[1,2,4],[6,7,3],[8,9,4]])
print(matriz)
#acessando elemento da matriz
print(matriz[0, 1])
```

```python
# fatiamento [inicial, final -1:passo]
array = np.arange(0,9)
print(array)

fat = array[2:6:1]
print(fat)

# matriz
matriz = np.array([[10, 20, 30], [40, 50, 60]])
print(matriz)
# acessar todas as linhas mas somente coluna 1 e 2
fat = matriz[:, 1:3]
print(fat)
```

```python
# indexação booleana
dados = np.arange(1, 11)
print(dados)

# criando uma condição
condicao = dados > 5
print(condicao)

elementos = dados[condicao]
print(elementos)
```

```python
# elementos pares e maiores que 5
condicao1 = dados % 2 == 0
condicao2 = dados > 5
elementos = dados[condicao1 & condicao2]
print(elementos)
```
