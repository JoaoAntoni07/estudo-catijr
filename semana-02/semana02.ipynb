### Operação com arrays

```python
# operacções aritmétricas
#som
array1 = np.array([1,2,3,4])
array2 = np.array([4,5,6,7])
array_aux = array1 + array2
print(array_aux)

#subtração
array1 = np.array([1,2,3,4])
array2 = np.array([4,5,6,7])
array_aux = array1 - array2
print(array_aux)

#produto
array1 = np.array([1,2,3,4])
array2 = np.array([4,5,6,7])
array_aux = array1 * array2
print(array_aux)

#divisão
array1 = np.array([1,2,3,4])
array2 = np.array([4,5,6,7])
array_aux = array1 / array2
print(array_aux)
```

```python
# operação com escalar
#soma
array_aux = array2 + 5
print(array_aux)

#subtração
array_aux = array2 - 5
print(array_aux)

#produto
array_aux = array2 * 5
print(array_aux)

#divisão
array_aux = array2 / 5
print(array_aux)
```

```python
# broadcasting
array1 = np.array([[10,11,13],[15,16,17]])
print(array1)
array2 = np.array([10,20,30])
print(array2)

soma = array1 + array2
print(soma)
```

### Indexação e fatiamento

```python
# fatiamento
# acessando elementos
array = np.arange(10, 60, 10)
print(array)

print(array[0])
print(array[2])

#matriz
matriz = np.array([[1,2,4],[6,7,3],[8,9,4]])
print(matriz)
#acessando elemento da matriz
print(matriz[0, 1])
```

```python
# elementos pares e maiores que 5
condicao1 = dados % 2 == 0
condicao2 = dados > 5
elementos = dados[condicao1 & condicao2]
print(elementos)
```

### Apply e map

```python
# criar o df
dados = {
    'Nome' : ['Leo', 'José', 'Marcelo'],
    'Idade' : [25, 31, 19],
    'Status' : ['ativo', 'inativo', 'ativo']
}
df = pd.DataFrame(dados)
print(df)
```

```python
# apply
def categoria_idade(idade):
    if idade < 30:
        return 'Jovem'
    else:
        return 'Adulto'

df['Categorai_idade'] = df['Idade'].apply(categoria_idade)
print(df)
```

```python
# map
mapeamento_status = {'ativo' : 'Ativo', 'inativo' : 'Inativo'}
df['Status_formatado'] = df['Status'].map(mapeamento_status)
print(df)
```

### Funções de agragação

```python
dados = {
    'Cidade' : ['São Paulo', 'Rio de Janeiro', 'São Paulo', 'Belo Horizonte', 'Rio de Janeiro', 'São Paulo'],
    'Vendas' : [100, 150, 120, 80, 200, 90],
    'Produto' : ['A', 'B', 'A', 'C', 'B', 'A']
}
df = pd.DataFrame(dados)
print(df)
```

```python
# agrupar por cidade e calcular soma das vendas por cidade
vendas_por_cidade = df.groupby('Cidade')['Vendas'].sum()
print(vendas_por_cidade)
```

```python
media_vendas = df.groupby(['Cidade', 'Produto'])['Vendas'].mean()
print(media_vendas)
```
