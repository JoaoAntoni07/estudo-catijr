## Teste da hipótese

### Teste realizado para validar ou refutar suposições sobre uma população com base em dados de amostra.
### Pergunta escolhida para realizar o teste:
- O valor da conta no jantar é estatisticamente diferente do valor da conta no almoço?
### Hipóteses matemáticas:
- $H_0$ (Hipótese Nula): As médias são iguais. Não há diferença.
- $H_1$ (Hipótese Alternativa): As médias são diferentes.
### Usaremos um Teste T para amostras independentes. Se o resultado do valor-p ($p$) for menor que 0.05, nós rejeitamos a hipótese nula e confirmamos que existe diferença!

## Métricas descritivas e de dispersão

### Métricas descritivas utilizadas: média e mediana.
### Métricas de dispersão utlizadas: variância e desvio padrão.

```python
# importando bibliotecas necessárias
import pandas as pd
import seaborn as sns
from scipy import stats
```

```python
# 1- criando o dataset
df = sns.load_dataset('tips')
print(f"Vizualizando as cinco primeiras linhas do dataset: ")
df.head()
```

```python
# 2 - métricas descritivas e de dispersão
# 2.1- métricas descritivas analisando a coluna 'total_bill'
media = df['total_bill'].mean()
mediana = df['total_bill'].median()
print(f"MÉTRICAS DESCRITIVAS: ")
print(f"Média: {media:.2f}")
print(f"Mediana: {mediana:.2f}")
```

```python
# 2.2- métricas de dispersão analisando a coluna 'total_bill'
variancia = df['total_bill'].var()
desvio_padrao = df['total_bill'].std()
print(f"MÉTRICAS DE DISPRESÃO: ")
print(f"Variância: {variancia:.2f}")
print(f"Desvio padrão: {desvio_padrao:.2f}")
```

```python
# 3- teste da hipótese
# 3.1- separando dados em almoço e janta
contas_almoco = df[df['time'] == 'Lunch']['total_bill']
contas_janta = df[df['time'] == 'Dinner']['total_bill']
```

```python
# 3.2- executando teste estatístico
# stats.ttest_ind - função disponibilizada pela biblioteca scipy
estatistica_t, valor_p  = stats.ttest_ind(contas_almoco, contas_janta)
print(f"RESULTADO DO TESTE DAS HIPÓTESES: ")
print(f"Valor-P: {valor_p:.5f}")
# decisão
if valor_p < 0.05:
    print(f"CONCLUSÃO: Rejeitamos a hipótese H0. Existe uma diferença significativa de vendas entre o almoço e a janta.")
else:
    print(f"CONCLUSÃO: Falhamos em rejeitar a hipótese H0. Não podemos afirmar que há diferença entre as vendas do almoço e da janta.")
```
