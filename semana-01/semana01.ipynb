### While

```python
# exemplo usando contador
i = 1

while i <= 10:
    print(i)
    i += 1
```

### While True, Break, Continue

```python
# while true
while True:
    senha = input('Digite sua senha: ')
    if senha == 'joao':
        print('Acesso liberado')
        break
```

```python
# break
i = 0

while i < 10:
    print(i)
    if i == 7:
        break
    i += 1
```

```python
# continue
i = 0

while i < 20:
    i += 1
    if i == 10:
        continue
    else:
        print(i)
```

## Estrutura de Dados

### Listas

```python
# tamanho de uma lista
cores = ['Amarelo', 'Azul', 'Vermelho']
print(len(cores))
```

```python
# inserindo dados em uma lista
idades = [18, 22, 25, 24]
print(idades)
idades.append(28)
print(idades)
```

```python
# inserindo dados em uma posição específica
cores.insert(0, 'Marrom')
print(cores)
```

```python
# juntando listas
nomes1 = ['Joao', 'Paulo']
nomes2 = ['Pedro', 'Marcela']

print(nomes1)
print(nomes2)

nomes1.extend(nomes2)
print(nomes1)
```

```python
# removendo dados de uma lista (remove)
nomes1.remove('Paulo')
print(nomes1)
```

```python
# removendo dados de uma lista (pop)
nomes1 = ['Pedro', 'Marcelo', 'Mário', 'Nadir']
nomes1.pop(1)
print(nomes1)
```

```python
# removendo dados de uma lista (del)
anos = [2002, 2204, 2015, 2019]
print(anos)

del anos[3]
print(anos)
```

```python
# limpando todos os itens de uma lista
print(anos)
anos.clear()
print(anos)
```

### Percorrendo e fatiando listas

```python
# percorrendo lista com for
anos = [2001, 2002, 2003, 2004, 2005]
print(anos)

for i in anos:
    print(i)
```

```python
# percorrendo lista com in range
for i in range(len(anos)):
    print(i)
```

```python
# usando while
contador = 0
while contador < len(anos):
    print(anos[contador])
    contador += 1
```

```python
# fatiamento de listas
compras = ['arroz', 'feijão', 'banana', 'alecrim']
print(compras)

print(compras[0:4:2])

print(compras[-1])
```

### Listas aninhadas

### Introdução a dicionários

```python
# sintaxe
carros = {}

pessoas = {
    'nome': 'João',
    'idade': 21,
    'altura': 1.83
}
print(pessoas)
```

```python
# acessando um item do dicionário
print(pessoas['nome'])
print(pessoas['idade'])
```

```python
# modificando item de dicionário
pessoas = {
    'nome': 'João',
    'idade': 21,
    'altura': 1.83
}
pessoas['altura'] = 1.84

print(pessoas)
```

```python
# tamanho de um dicionário
print(len(pessoas))
```

```python
# acessando valores de um dicionário (get)
pessoas = {
    'nome': 'João',
    'idade': 21,
    'altura': 1.83
}
print(pessoas.get('idade'))
```

```python
# retornando chaves de um dicionário
print(pessoas.keys())
```

```python
# retornando valores de um dicionário
print(pessoas.values())
```

```python
# retorna chave e valor
print(pessoas.items())
```

```python
# verificando a existência de uma chave específica
'nome' in pessoas
```

```python
# alterando valores no dicionário
pessoas['idade'] = 22
print(pessoas)
```

```python
# adicionando nova chave/valor no dicionário
pessoas['cpf'] = 1023032
print(pessoas)
```

```python
# removendo um item chave
pessoas.pop('altura')
print(pessoas)
```

```python
# removendo o último elemento inserido
pessoas.popitem()
print(pessoas)
```

```python
# utilizando del
pessoas = {
    'nome': 'João',
    'idade': 21,
    'altura': 1.83
}

del pessoas['altura']
print(pessoas)
```
