### List Comprehension

```python
# sintaxe
[expressao for item in interavel]
```

```python
# criando lista com quadrados dos números de 1 a 10
# lista
lista_quadrados = []

for i in range(1, 11):
    lista_quadrados.append(i**2)

print(lista_quadrados)

#list comprehension
quadrados = [i**2 for i in range(1, 11)]
print(quadrados)
```

```python
# lista de números pares
lista_pares = []

for i in range(1, 21):
    if (i%2) == 0:
        lista_pares.append(i)

print(lista_pares)

# list comprehension
pares = [i for  i in range(1, 21) if i % 2 == 0]
print(pares)
```

```python
# strings
frase =  'sou corinthiano maloqueiro e sofredor'

lista_palavras = [palavra.upper() for palavra in frase.split()]
print(lista_palavras)
```
